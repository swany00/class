- 피보나치 수열 생성
$$f_1=f_2=1, ~f_i=f_{i-1}+f_{i-2}, ~i=3,4,\ldots$$

```r
f <- numeric(length = 10)
# 길이가 10이고 실수형인 벡터를 설정
f[1] <- 1
f[2] <- 1
for (i in 3:10) { f[i] <- f[i-1] + f[i-2] }
# loop 문을 통해서 피보나치 수열을 계산
print(f)
```

- 재사용규칙 : 길이가 다른 벡터간에 연산을 실시하면 길이가 짧은 벡터를
반복적으로 다시 사용해서 연산을 실시하는 재사용규칙이 적용된다. 여기에서 긴 벡터의 길이를 짧은 벡턱의 길이로 나누었을 때, 정수가 되는 것이 기본 설정이다.

```r
a <- 1:6 #(1,2,3,4,5,6)
b <- 2:4 #(2,3,4)
print(a)
print(b)
print(a+b) #(1,2,3,4,5,6)+(2,3,4,2,3,4)
```

```r
a <- 1:7
b <- 2:4
print(a)
print(b)
print(a+b)
```

```r
print(a+10)
print(a*10)
```

```r
x <- 1:5
print(x)
print(x>3) ; print(sum(x>3))
```

```r
print(x>4) ; print(x<=3)
print((x>4)&(x<=3))
print((x>4)|(x<=3))
```

### 2.4 특정 형태의 벡터 생성 함수
- 특정 패턴을 가지는 벡터를 생성해 내는 함수가 있다
- 예: seq()는 등차수열을 생성한다.

```r
g = seq(from = 0, to = 10, by = 2.5)
print(g)
```

```r
g = seq(0, 10, length = 4)
print(g)
```

```r
g = seq(10,1)
print(g)
```

- 예: rep()는 값이 반복되는 벡터를 생성한다.

```r
g = rep(0, times = 10)
print(g)
g = rep(0:1, 3)
print(g)
```

```r
g = rep(c(1, 2, 3), 2)
print(g)
```

```r
g = rep(c(1, 2, 3), c(1, 2, 3))
print(g)
g = rep(c(1,2,3), each=4)
print(g)
```

### 2.6 필터링
- 특정한 조건에 부합하는 값만 남기는 기법이다
- 벡터의 일부만을 추출해 낸다는 측면에서는 인덱싱과 유사하다
-  예: 5를 초과하는 값들만 추출한다

```r
z <- c(1, 8, 9, 2, 7, 10, 5, 6, 4, 3)
print(z[z>5])
print(z[z<=6])
```

```r
print(z>5)
print(sum(z>5))
print(mean(z>5))
```

```r
index = which(z>5)
print(index)
print(z[index])
```

- ifelse(test, yes, no)는 test가 TRUE이면 yes를 값으로
출력하고 FALSE이면 no를 값으로 출력한다.
- 여기서 test, yes, no는 모두 벡터이다.
- 예: x > 0이면 2*x를 출력하고 x ≤ 0이면 x를 출력하시오.

```r
x = c(-0.6, 0.2, -0.8, 1.6, 0.3, -0.8, 0.5, 0.7, 0.6, -0.3)
ind1 = which(x<=0)
print(ind1)
x[ind1] = 2*x[ind1]
print(x)
```

```r
x <- c(-0.6, 0.2, -0.8, 1.6, 0.3, -0.8, 0.5, 0.7, 0.6, -0.3)
print(ifelse(x>0, 2*x, x))
```
